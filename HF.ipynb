```python
from transformers import pipeline

# Load a question-answering pipeline
Youtubeer = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Input for Question Answering
question = "What is the future of AI in medicine?"
context = "Artificial intelligence is rapidly transforming various sectors, and medicine is no exception. The future of AI in medicine is expected to revolutionize diagnostics, drug discovery, personalized treatment plans, and even robotic surgery, leading to more efficient and effective healthcare."

# Output
result = Youtubeer(question=question, context=context)

print(f"Question: {question}")
print(f"Answer: {result['answer']}")
print(f"Score: {result['score']:.4f}")
```

```python
from transformers import pipeline

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Input
article = """
Machine learning is a method of data analysis that automates analytical model building.
It is a branch of artificial intelligence based on the idea that systems can learn from data,
identify patterns and make decisions with minimal human intervention.
"""

# Output
summary = summarizer(article, max_length=50, min_length=20, do_sample=False)

print(summary[0]['summary_text'])
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Input
sentence1 = "What is the capital of France?"
sentence2 = "Paris is the capital city of France."

# Output
embedding1 = model.encode(sentence1, convert_to_tensor=True)
embedding2 = model.encode(sentence2, convert_to_tensor=True)

similarity = util.pytorch_cos_sim(embedding1, embedding2)

print(f"Similarity score: {similarity.item():.4f}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# 1. Input prompt
prompt = "Once upon a time"
input_ids = tokenizer(prompt, return_tensors="pt").input_ids

# 2. Generate continuation
output_ids = model.generate(input_ids, max_length=50, do_sample=True, top_k=50)

# 3. Decode result
generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(generated_text)
```

```python
from transformers import pipeline

multi_label = pipeline("zero-shot-classification", model="joeddav/xlm-roberta-large-xnli")

text = "I enjoy hiking, swimming, and learning new languages."
candidate_labels = ["sports", "travel", "education", "leisure", "health"]

result = multi_label(text, candidate_labels, multi_label=True)

for label, score in zip(result['labels'], result['scores']):
    print(f"{label}: {score:.4f}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# Load tokenizer and model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Text input
text = "I love Hugging Face and NLP!"

# Tokenize
inputs = tokenizer(text, return_tensors="pt")  # pt = PyTorch

# Run model
with torch.no_grad():
    logits = model(**inputs).logits
    probs = F.softmax(logits, dim=-1)

# Decode prediction
predicted_class = probs.argmax().item()
label_map = model.config.id2label

print(f"Predicted class: {label_map[predicted_class]}, Score: {probs[0][predicted_class]:.4f}")
```
